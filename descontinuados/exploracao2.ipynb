```python
import pandas as pd
```

```python
dados = pd.read_parquet('data/dados.parquet')
```

```python
dados['date_purchase'] = pd.to_datetime(dados['date_purchase'], format='%Y-%m-%d')
```

```python
dados.head()
```

```python
from datetime import datetime

today = datetime.now()
print(today)
df_rfv = dados.groupby('fk_contact').agg(
    recency = ('date_purchase', lambda date: (today - date.max()).days),
    frequency = ('date_purchase', 'count'),
    monetary = ('gmv_success', 'sum'),
    ticket_medio = ('total_tickets_quantity_success', 'mean')
).reset_index()
df_rfv
```

```python
#features = ['recency','frequency','monetary', 'ticket_medio']
features = ['recency','frequency','monetary']

df_train = df_rfv[features].copy()
```

- Normalizar

```python
from sklearn.discriminant_analysis import StandardScaler


scaler = StandardScaler()
df_normalizado = pd.DataFrame(scaler.fit_transform(df_train), columns=features)
df_normalizado.head()
```

```python
df_normalizado.corr()
```

```python
from sklearn.cluster import KMeans


modelo_kmeans = KMeans(3, random_state=42, n_init='auto')
clusters = modelo_kmeans.fit_predict(df_normalizado)
```

```python
clusters
```

```python
df_rfv['clusters_kmeans'] = clusters 
```

```python
df_rfv
```

```python
df_normalizado.head()
```

```python
import hdbscan

modelo_hbscan = hdbscan.HDBSCAN(min_samples=8)

clusters = modelo_hbscan.fit_predict(df_normalizado)
```

```python
df_rfv['clusters_hdbscan'] = clusters
df_rfv
```

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np
import matplotlib.pyplot as plt

# Suponha que `df_padronizado` é o seu DataFrame com os dados numéricos já padronizados
# Suponha que min_samples = 5s

# k é igual a min_samples
k = len(features) * 2

# Crie uma instância de NearestNeighbors
# O algoritmo 'brute' é simples e eficaz para conjuntos de dados menores
nbrs = NearestNeighbors(n_neighbors=k).fit(df_normalizado)

# Calcule as distâncias dos k vizinhos mais próximos
distancias, indices = nbrs.kneighbors(df_normalizado)

# Ordene as distâncias do k-ésimo vizinho (última coluna do array de distâncias)
distancias_ordenadas = np.sort(distancias[:, k-1], axis=0)



# Plote o gráfico
plt.figure(figsize=(10, 6))
plt.plot(distancias_ordenadas)
plt.xlabel("Pontos de Dados (ordenados)")
plt.ylabel(f"Distância para o {k}-ésimo vizinho")
plt.title("Gráfico de Distância K-ésima para o DBSCAN")

# Ajuste os limites dos eixos aqui
plt.ylim(0, 0.04) # Ajusta o eixo Y para ir de 0 a 5
#plt.xlim(0, 100) # Ajusta o eixo X para ir de 0 a 100, se necessário

plt.grid(True)
plt.show()
```

```python
# Calcule a diferença entre os pontos para encontrar o maior salto
diferencas = np.diff(distancias_ordenadas)
maior_salto = np.argmax(diferencas)
print(f"O ponto com o maior salto de distância é o índice {maior_salto}")
print(f"O valor de EPS correspondente é: {distancias_ordenadas[maior_salto]}")
```
